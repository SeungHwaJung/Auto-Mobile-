# 📘 YOLO 용어 정리

| 용어 | 설명 |
|------|------|
| YOLO | You Only Look Once - 실시간 객체 탐지 알고리즘 |
| Bounding Box | 객체를 둘러싼 사각형 (x, y, width, height) |
| Confidence Score | 해당 클래스일 확률 |
| NMS (Non-Maximum Suppression) | 겹치는 박스를 제거해 최종 객체 선택 |
| COCO Dataset | 객체 탐지 학습용 데이터셋 (80개 클래스) |
| mAP | Mean Average Precision - 정확도 지표 |
| Anchor Box | 다양한 크기의 객체 탐지를 위한 초기 기준 박스 |
| IoU | Intersection over Union - 예측과 실제 박스의 겹침 정도 |
| Pretrained Model | 사전 학습된 모델, fine-tuning 가능 |
| ONNX | 모델을 다른 프레임워크와 호환되도록 저장하는 포맷 |

```python
!pip install ultralytics
from google.colab import files
from ultralytics import YOLO # COCO 사전 훈련된 YOLOv8n 모델 로드
model = YOLO("yolov8n.pt") # 모델 정보 표시 (선택사항)
model.info() # COCO8 예제 데이터셋으로 100 에포크 훈련
results = model.train(data="coco8.yaml", epochs=10, imgsz=640) # 사진 업로드하고 경로 설정
uploaded = files.upload()
image_path = list(uploaded.keys())[0] # 업로드한 이미지에 대해 YOLOv8n 모델로 추론 실행
results = model(image_path)
results[0].show()
```
